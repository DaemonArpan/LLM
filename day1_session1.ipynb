# sentiment analysis using Transformer model

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier (
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!"
    ]
)
```

# text Generation model lile chatGPT

```python
generator = pipeline("text-generation")
generator(
    "I love the Jurassic Park movies! I think the original Jurassic Park is a masterpiece. ",
    max_length=50,
    num_return_sequences=1,
)
```

# QnA Models

```python
qa = pipeline("question-answering")
qa (
    question="Who is the protagonist of Jurassic Park?",
    context = "Dr. Allel Grant saves the crew using his 3D printed Velociraptor SoundBox...",
)
```

# Language Translator

```python
# Language Translator
translator = pipeline("translation_en_to_de")
translator(" random text generator!")
```
